```python
def predict_desease(arr):  
    
    ar = np.zeros(len(x.columns))
    for i in arr:
        loc_index = np.where(x.columns==col_dict[i])[0][0]
        if loc_index >= 0:
            ar[loc_index] = 1
            
    return disease_dict[model.predict([ar])[0]]
```
